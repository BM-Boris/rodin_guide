## Statistical Tests

### One Way Anova

```python
obj.oneway_anova('Treatment')
```

```python
obj_sl = obj[obj.features[obj.features['p_adj(owa) Treatment']<0.01]]
obj_sl.run_umap()

obj_sl.plot(hue='Treatment',title=f'UMAP - {obj_sl.features.shape[0]} metabolites (p_value<0.01)')
```

```python
obj_sl.clustergram(title=f'Clustergram - {obj_sl.features.shape[0]} metabolites (p_value<0.01)',hue='Treatment')
```

### PLS DA

```python
obj.pls_da('Treatment')
```

```python
obj_sl = obj[obj.features[obj.features['vips']>2.25]]
obj_sl.run_umap()

obj_sl.plot(hue='Treatment',title=f'UMAP - {obj_sl.features.shape[0]} metabolites (vip>2.25)')
```

```python
obj_sl.clustergram(title=f'Clustergram - {obj_sl.features.shape[0]} metabolites (vip>2.25)',hue='Treatment')
```

## PATHWAY Analysis

```python
obj.fold_change('Treatment',reference='VA_0')
obj.analyze_pathways('p_adj(owa) Treatment','lfc (VA_0 vs others)',mode='f_positive')[:6]
```

```python
obj.show_compounds(cutoff_path=0.001)
```
